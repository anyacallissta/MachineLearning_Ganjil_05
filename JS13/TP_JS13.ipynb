```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

# 1. Load dataset MNIST
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalisasi data (0-255 → 0-1)
X_train = X_train / 255.0
X_test = X_test / 255.0

# One-hot encoding label
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# 2. Bangun model JST dengan 2 hidden layer
model = Sequential([
    Flatten(input_shape=(28, 28)),          # ubah 28x28 → 784
    Dense(128, activation='relu'),          # hidden layer 1
    Dense(64, activation='relu'),           # hidden layer 2
    Dense(10, activation='softmax')         # output layer (10 kelas)
])

# 3. Kompilasi model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# 4. Latih model
model.fit(X_train, y_train, epochs=5, batch_size=32, verbose=1)

# 5. Evaluasi model
loss, acc = model.evaluate(X_test, y_test)
print(f"Akurasi pada data uji: {acc:.4f}")
```

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
import time
import pandas as pd

# 1. Load dataset MNIST
(X_train, y_train), (X_test, y_test) = mnist.load_data()

X_train = X_train / 255.0
X_test = X_test / 255.0

y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# 2. Fungsi Membangun Model
def build_model(activation="relu"):
    model = Sequential([
        Flatten(input_shape=(28,28)),
        Dense(256, activation=activation),
        Dense(128, activation=activation),
        Dense(64, activation=activation),  # hidden layer tambahan
        Dense(10, activation="softmax")
    ])
    model.compile(optimizer="adam",
                  loss="categorical_crossentropy",
                  metrics=["accuracy"])
    return model

# 3. Daftar eksperimen
experiments = [
    ("ReLU", "relu"),
    ("Sigmoid", "sigmoid")
]

results = []

# 4. Jalankan eksperimen
for name, act in experiments:
    print(f"\nTraining model dengan aktivasi: {name}")

    model = build_model(activation=act)

    start = time.time()
    history = model.fit(X_train, y_train, epochs=5, batch_size=32, verbose=0)
    end = time.time()

    loss, acc = model.evaluate(X_test, y_test, verbose=0)

    results.append({
        "Aktivasi": name,
        "Akurasi Test": acc,
        "Waktu Training (detik)": end - start
    })

# 5. Tampilkan hasil dalam tabel
df_results = pd.DataFrame(results)
print(df_results)
```
